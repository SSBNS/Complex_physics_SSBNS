```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as random
```

```python
# 1.4
trials = 10**6
line = np.ones(100)
for i in range(trials):
    i,j = random.randint(0,100,2)
    A , B = line[i],line[j]
    sum = A+B
    A_frac = random.randint(0,sum+1)
    line[i] = A_frac
    line[j] = sum-A_frac
    if sum-A_frac < 0:
        print("error")
        break
```

```python
plt.hist(line,density=True,bins=np.arange(0,5))
# x = np.linspace(0,5)
# plt.plot(x,1/2**x*0.4)
plt.show()
```

```python
# 1.3
N = 10**4
trials = 10**5
Pop = np.zeros(N)
mean = []
mean_dif = []
for i in range(trials):
    Pop_i = np.copy(Pop)
    select = random.choice([True,False],N)
    Pop[select]+=1
    Pop[~select]=0
    mean_dif.append(np.mean(np.sqrt((Pop_i-Pop)**2)))
    mean.append(np.mean(Pop))
```

```python
print(max(Pop))
bins=np.arange(0,max(Pop))
plt.hist(Pop,density=True,bins=bins)
plt.plot(np.arange(0,18),1/2**(np.arange(0,18))/2)
plt.show()
plt.plot(np.arange(trials),mean_dif)
#plt.ylim(0,10**-1)
plt.show()
plt.plot(np.arange(trials),mean)
plt.show()
print(np.mean(Pop))
```

```python
# 1.9
def energy(grid,J):
    up = grid*np.roll(grid,1,0)
    down = grid*np.roll(grid,-1,0)
    left = grid*np.roll(grid,1,1)
    right = grid*np.roll(grid,-1,1)
    E_grid = up+down+left+right
    E = -1/2*J*np.sum(E_grid)
    return E

def spinflip(grid):
    i,j = random.randint(0,np.shape(grid)[0],2)
    flip_grid = np.copy(grid)
    flip_grid[i,j] *= -1
    return flip_grid

def exp_factor(delta_E,T):
    x = np.exp(-delta_E/T)
    return x

def select(delta_E,T):
    a= 0
    if delta_E<0:
        return 1
    else:
        random_number = random.rand()
        exp_number = exp_factor(delta_E,T)
        if random_number < exp_number:
            return 1
        else: 
            return 0

```

```python
T=4
n_max=10**6
n_0 = n_max/4
N_site=20
grid = random.randint(0,2,size=[N_site,N_site])
grid = 2*grid-1
J=-1
```

```python
E = []
s = []
for l in range(int(n_max)):
    prev_grid = np.copy(grid)
    new_grid = spinflip(grid)
#    prev_subgrid = grid[i-1:i+2,j-1:j+2]
 #   new_subgrid = new_grid[i-1:i+2,j-1:j+2]
  #  prev_grid = np.copy(grid)
   # new_grid = spinflip(grid)[-1]
    delta_E = energy(new_grid,J) - energy(prev_grid,J)
    selection = select(delta_E,T)
    if selection==1:
        grid=new_grid
    else:
        grid = prev_grid
    if l > n_0:
        E.append(energy(grid,J))
        s.append(np.sum(grid))
```

```python
x = np.arange(len(E))
plt.plot(x,E,label= "energy")
plt.plot(x,s,label="moment")
plt.legend()
plt.show()
```
